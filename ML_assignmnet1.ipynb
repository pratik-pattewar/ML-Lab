```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(url)
```

```python
# Dataset shape and overview
print("Dataset Shape:", df.shape)
print("\nFirst 5 rows of dataset:")
print(df.head())
```

```python
# Dataset info
print("\nDataset Info:")
print(df.info())
```

```python
# Check for missing values
print("\nMissing Values in Dataset:")
print(df.isnull().sum())
```

```python
# Statistical summary
print("\nStatistical Summary:")
print(df.describe())
```

```python
# Unique values in species column
print("\nUnique Species Classes:", df['species'].unique())
```

```python
# Value counts
print("\nSpecies Value Counts:")
print(df['species'].value_counts())
```

```python
# Correlation matrix
print("\nCorrelation Matrix:")
print(df.corr(numeric_only=True))
```

```python
# Plot heatmap of correlation
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pairplot for feature relationships
sns.pairplot(df, hue='species')
plt.suptitle('Pairplot of Features', y=1.02)
plt.show()
```

```python
# Group by species and get mean
print("\nGroup By Species - Mean Values:")
print(df.groupby('species').mean())
```

```python
# Sorting data based on petal_length
sorted_df = df.sort_values(by='petal_length', ascending=False)
print("\nTop 5 rows with highest petal length:")
print(sorted_df.head())
```

```python
# Filtering example: Sepal Length > 6
filtered_df = df[df['sepal_length'] > 6]
print("\nRows where Sepal Length > 6:")
print(filtered_df.head())
```

```python
# Boxplot for visualizing distributions
plt.figure(figsize=(8,6))
sns.boxplot(x='species', y='sepal_length', data=df)
plt.title('Boxplot of Sepal Length by Species')
plt.show()
```
